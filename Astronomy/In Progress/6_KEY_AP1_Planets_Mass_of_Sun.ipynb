Convert units as needed.

```
data['T']=data['Orbital_period']*3.154*10**7 # years into s
data['R'] = data['Distance_from_Sun']*1.496*10**11 # Fill in conversion from AU into m
data['v'] = data['Orbital_speed']*1000 # Fill in conversion from km/s into m/s
```

Define "G" and other necessary constants.

```
G=6.67*10**-11
C = 4*math.pi**2/G
```

Write code to create new columns with data to graph and check on your data. One sample is done for you. Use your answers from above to determine the data you need.

```
data['T^2'] = data['T']**2
data['R^3'] = data['R']**3
data['v^2'] = data['v']**2
data['1/R'] = data['R']**(-1)
data['CR^3'] = data['R^3']*C
data['G/R'] = data['1/R']*G
data.head(3)
```

Scrolling to the right on the table above will let you view your new data columns.

Plot the data such that the slope of the line of best fit is the mass of the sun.

```
# Write code here.
```

Three samples are provided below.

```
x=data['T^2']
y=data['CR^3']
plt.scatter(x,y)
z = np.polyfit(x,y, 1) #makes linear regression
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.xlabel('Period squared (s^2)')
plt.ylabel('C x Distance from Sun cubed(kg s^2)')
plt.title('Mass of Sun')
plt.grid(True) #shows gridlines
plt.show()
slope = z[0] #sets the slope of the line = to a variable called slope
print("Mass of Sun =",slope,"kg")
```

```
# line of best fit parameters
xmin = 0  # lower bound for your plot
xmax = 10**-21 # upper bound for your plot
slope = 2*10**30 # est of slope (will be very large, use sci. not.)guess and check until you are satisfied
x = np.arange(xmin, xmax, (xmax-xmin)/200) # makes a series of equally spaced x-values
y = slope * x # calculates the y-values

# make the plot
fig, ax = plt.subplots(1,1, figsize=(6,4))
plt.plot(data['G/R'],data['v^2'], linestyle = ' ', marker='o', color='b', markersize=10)
ax.plot(x, y, color='r')  # plots the curve
plt.xlabel('G/Distance from Sun (m^2/kg s^2)')
plt.ylabel('Orbital Speed squared (m^2/s^2)')
plt.title('Mass of Sun')
plt.grid(True) #shows gridlines
plt.show()
```
