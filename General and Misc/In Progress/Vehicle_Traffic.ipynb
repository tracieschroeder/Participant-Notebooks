# Weekly Vehicle Traffic Volume

A traffic engineer is looking to optimize vehicle traffic on a busy highway. The graphs below show the percent change in traffic volume for several weeks: one for all vehicles, one for passenger vehicles, and one for trucks. The last one shows all the data collected as one graph.

Answer the questions that follow each graph, and then use your responses to determine the best course of action.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://data.transportation.gov/api/views/yeig-3uz6/rows.csv?accessType=DOWNLOAD')

week = df['Week']
vehicle = df['% Change All Vehicles']

bar_width = .6
x = np.arange(len(week))

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.bar(x, vehicle, bar_width, color = 'mediumaquamarine', label = 'Vehicle')

plt.xlabel('Week')
plt.ylabel('% Change All Vehicles')
plt.title('Weekly Traffic Volume')

plt.grid(False)
plt.axis([0, 175, -50, 10])

plt.show()
```

```python
df = pd.read_csv('https://data.transportation.gov/api/views/yeig-3uz6/rows.csv?accessType=DOWNLOAD')

week = df['Week']
vehicle = df['% Change All Vehicles']
passenger = df['% Change Passenger']
truck = df['% Change Truck']

bar_width = .5
x = np.arange(len(week))

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.bar(x, vehicle, bar_width, color = 'mediumaquamarine', label = 'Vehicle')
plt.bar(x + bar_width, passenger, bar_width, color = 'cornflowerblue', label = 'Passenger')
plt.bar(x + bar_width, truck, bar_width, color = 'darkorange', label = 'Truck')

plt.xlabel('Week')
plt.ylabel('% Change')
plt.title('Weekly Traffic Volume')

plt.grid(False)
plt.axis([0, 175, -60, 30])

plt.legend(['Vehicle', 'Passenger', 'Truck'])
plt.show()
```
