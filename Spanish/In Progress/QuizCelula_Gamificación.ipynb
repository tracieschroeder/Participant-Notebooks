```python
#1. Define las preguntas y respuestas:
#Crea una lista de preguntas y sus respectivas respuestas.
#Cada pregunta debe tener una respuesta correcta y varias respuestas incorrectas.
# Una Lista de Diccionarios: un diccionario por cada pregunta
preguntas = [
    {
        "pregunta": "¿Cuál de las siguientes opciones describe mejor a las células procariotas?",
        "respuestas": ["Tienen un núcleo definido",
                       "Carecen de membrana celular",
                       "Contienen organelos membranosos",
                       "Su material genético está disperso en el citoplasma"],
        "respuesta_correcta": "Su material genético está disperso en el citoplasma"
    },
    {
        "pregunta": "¿Cuál de las siguientes opciones es una característica de las células eucariotas?",
        "respuestas": ["Carecen de membrana plasmática",
                       "Tienen un núcleo definido",
                       "No contienen organelos",
                       "Su material genético está disperso en el citoplasma"],
        "respuesta_correcta": "Tienen un núcleo definido"
    },
    {
        "pregunta": "¿Cuál de las siguientes opciones describe mejor a las células vegetales?",
        "respuestas": ["Carecen de pared celular",
                       "Tienen cloroplastos para la fotosíntesis",
                       "No tienen vacuolas",
                       "Su membrana celular es flexible y no rígida"],
        "respuesta_correcta": "Tienen cloroplastos para la fotosíntesis"
    },
      {
        "pregunta": "¿Cuál de las siguientes opciones es una característica de las células animales?",
        "respuestas": ["Contienen organelos",
                       "Contienen cloroplastos para la fotosíntesis",
                       "No tienen mitocondrias",
                       "Su membrana celular es rígida y no flexible"],
        "respuesta_correcta": "Contienen organelos"
    },
      {
        "pregunta": "¿Cuál de las siguientes opciones describe mejor a las células procariotas?",
        "respuestas": ["Tienen un núcleo definido",
                       "Carecen de membrana celular",
                       "Contienen organelos membranosos",
                       "Su material genético está disperso en el citoplasma"],
        "respuesta_correcta": "Su material genético está disperso en el citoplasma"
    },
    # Agrega más preguntas aquí
]
#preguntas
```

```python
#2. Importa el módulo random: Necesitarás el módulo random
# para mezclar las respuestas de cada pregunta.
import random
```

```python
#3. Mezcla las respuestas: Utiliza la función random.shuffle()
#para mezclar las respuestas de cada pregunta.

for pregunta in preguntas:
    random.shuffle(pregunta["respuestas"])
```

```python
#4. Presenta las preguntas y opciones de respuesta: Utiliza un bucle for
#para recorrer la lista de preguntas y mostrarlas junto con las opciones
#de respuesta.
for i, pregunta in enumerate(preguntas):
    print(f"Pregunta {i+1}: {pregunta['pregunta']}")
    for j, respuesta in enumerate(pregunta["respuestas"]):
        print(f"{j+1}. {respuesta}")
    print()
```

```python
#5. Solicita la respuesta del usuario: Pide al usuario que ingrese el número
#de la opción de respuesta que cons
respuestas_usuario = []
for i, pregunta in enumerate(preguntas):
    print(f"Pregunta {i+1}: {pregunta['pregunta']}")
    for j, respuesta in enumerate(pregunta["respuestas"]):
        print(f"{j+1}. {respuesta}")
    respuesta_usuario = input("Ingrese el número de la respuesta correcta: ")
    respuestas_usuario.append(respuesta_usuario)

```

```python
#6. Verifica las respuestas: Compara las respuestas del usuario
#con las respuestas correctas y muestra el puntaje obtenido.
puntaje = 0
for i, pregunta in enumerate(preguntas):
    if respuestas_usuario[i] == pregunta["respuesta_correcta"]:
        puntaje += 1

#print(f"Tu puntaje es: {puntaje}/{len(preguntas)}")
vTC=puntaje;
vTI=len(preguntas)-vTC;
print("Cantidad de Respuestas Correctas:",vTC)
print("Porciento de Respuestas Correctas:",round((vTC/len(preguntas)*100),0),'%')
print("Cantidad de Respuestas Incorrectas:",vTI)
print("Porciento de Respuestas Incorrectas:",round((vTI/len(preguntas)*100),0),'%')
```
