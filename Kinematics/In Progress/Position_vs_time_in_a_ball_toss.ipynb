# **Part 4:Calculating velocity from height**
Velocity is defined as the change in position over a set amount of time.
Velocity can be calculated from the equation v=(final height-initial height)/(final time-initial time)
We will use a code to calculate velocity from the height and time data .

```python
num_rows = data.shape[0] #This tells us how many rows are in the data file
height_diffs = [] #This creates a column in the file for height difference
time_diffs = [] #This creates a column in the file for time difference

i=0
while (i < num_rows - 1): # Iterate up to num_rows - 1
  height_difference = data['DataSet 1:Y(m)'][i+1] - data['DataSet 1:Y(m)'][i] #This subtracts one height from the previous height
  time_difference = data['DataSet 1:Time(s)'][i+1] - data['DataSet 1:Time(s)'][i] #This subtracts one time from the previous time
  height_diffs.append(height_difference)
  time_diffs.append(time_difference)
  i += 1

data = data.iloc[:-1] #This removes the last row of data, which does not have a final height or time to subtract it

data['height_difference'] = height_diffs
data['time_difference'] = time_diffs

# Now lets look at our data and see if it looks right. Ignore the SettingWithCopyWarnings.
data.head(15)
```

Does the data look right to you? Check by subtracting the first two times and the first two Y (m).

Now lets calculate the velocity. Alter the code below to make sure you calculate velocity correctly using the equation velocity=(height difference)/(time difference)


```python
velocity=data['height_difference']/data['time_difference']
data['velocity']=velocity
plt.scatter(data['DataSet 1:Time(s)'],data['velocity'], marker="o",color='green')
#These lines will create labels on the x axis, y axis, and the title.
#Change the red words in the next three rows to reflect the data.
plt.xlabel('time')
plt.ylabel('velocity')
plt.title('Age vs. Shoe Size')
```

# **Part 5:Finding the slope of the line**
We could have the computer find the best fit line like we did up in Part 2. But sometimes it is more useful to fit the line yourself.

```python
#This code will create a list of x values between a minimum and a maximum.
#Change the c to the minimum time in the data.
#Change the d to the maximum time in the data.
x = np.linspace(1.4, 1.9, num=120)

#This code will substitute the x values below into the y equation defined by the expression below.
#In the code below, change the letter m to a number you predict will be the slope of the line. Change the letter b to a number you predict will be the y intercept of the line.
#If your number does not create a good fit, try again until you get it!
y = -9.8*x + 18.5

plt.plot(x, y, color='r')
plt.scatter(data['DataSet 1:Time(s)'],data['velocity'], marker="o",color='green')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time of a ball moving upwards')
plt.show()
```
